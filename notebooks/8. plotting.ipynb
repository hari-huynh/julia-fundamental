# Plotting

## Basics
There are a few different ways to plot in Julia (including calling PyPlot)

Here we'll show you how to use `Plots.jl`

```julia
# Pkg.add("Plots")
using Plots
```

One of the advantages to `Plots.jl` is that it allows you to seamlessly change backends. In this notebook, we'll try out the `gr()` and `plotlyjs()` backends.

```julia
x = -3:0.1:3
f(x) = x^2

y = f.(x)
```

**Now let's load the GR backend**

```julia
gr()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

**Without changing syntax, we can try this with the `plotlyjs()` backend**

```julia
plotlyjs()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

## Getting slightly fancier

```julia
globaltemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numpirates = [45000, 20000, 15000, 5000, 400, 17]
```

```julia
# First, plot the data
plot(numpirates, globaltemperatures, legend=false)
scatter!(numpirates, globaltemperatures, legend=false)

# This reverses x axis we can see how the temperature change
xflip!()

# Add titles and labels
xlabel!("Number of Pirates [Approximate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
```

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)

plot(p1, p2, p3, p4, layout=(2,2), legend=false)
```
